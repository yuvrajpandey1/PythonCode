Sets are built-in data type in Python used to store collections of unique items. They are unordered, meaning that the elements do not follow a specific order, and they do not allow duplicate elemets. Sets are useful for membership tests, eliminating duplicate entries, and perfomring mathematical set operations like union, intersection, difference, and symmetric difference.

```python
## create a set
my_set = {1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set = set()
print(type(my_empty_set))
```

```python
my_set=set([1,2,3,4,5,6])
print(my_set)
```

```python
my_empty_set = set([1,2,3,6,5,4,5,6])
print(my_empty_set)
```

```python
# adding sets value

my_set.add(7)
print(my_set)
my_set.add(7)
print(my_set)
```

```python
my_set.remove(3)
print(my_set)
```

```python
my_set.remove(10)
```

```python
my_set.discard(11)
print(my_set)
```

```python
## pop method
removed_element = my_set.pop()
print(removed_element)
print(my_set)
```

```python
## Mathematical Operation

set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

## Union
union_set = set1.union(set2)
print(union_set)
```

```python
## Intersection
intersection_set=set1.intersection(set2)
print(intersection_set)
```

```python
### Counting Unique words in text

text = 'In this tutorial we are discussing about sets'
words=text.split()

## convert list of words to set to get unique words
unique_words = set(words)
print(unique_words)
print(len(unique_words))
```
