Dictionaries are unordered collections of items. They store data in key-value pairs. Keys must be unique and immutable (e.g. string, numbers, or tuples), while value can be of any type.

```python
## Creating Dictionaries
empty_dict = {}
print(type(empty_dict))
```

```python
empty_dict = dict()
print(type(empty_dict))
```

```python
student={
    'name' : 'Yuvraj',
    'age' : 25,
    'grade' : 24
    }
print(student)
```

```python
## accessing Dictionary Elements
student = {
    'name' : 'Yuvraj',
    'age' : 32,
    'grade' : 'A'
}
print(student)
```

```python
print(student['grade'])
print(student['age'])
```

```python
print(student.get('age'))
print(student.get('last_name'))
print(student.get('contact_number',"Not available"))
```

Dictionary are mutuable, so you can add, update or delete elements

```python
student['age'] = 35
print(student)
student['address'] = 'India'
print(student)
```

```python
# Dictionary methods

keys = student.keys()
print(keys)
values = student.values()
print(values)

items = student.items()
print(items)
```

```python
# shallow copy provides different memory location

student_copy1 = student.copy()
print(student_copy1)
print(student)
```

```python
student['name'] = 'Yuvraj1'
print(student_copy1)
print(student)
```

```python
## Iterating over dictionaries, we can use loop to iterate over dict

for keys in student.keys():
    print(keys)
```

```python
for values in student.values():
    print(values)
```

```python
for key,value in student.items():
    print(f'{key}:{value}')
```

```python
## Nested dictionaries

students = {
    'student1':{'name':'Yuvraj', 'age' :24},
    'student2':{'name':'Pandey', 'age' :26},
}
print(students)
```

```python
## Access nested dictionaries elements
print(students['student2']['name'])
print(students['student2']['age'])
```

```python
## Iterating over nested dictionaries

for student_id,student_info in students.items():
    print(f'{student_id}:{student_info}')
    for key,value in student_info.items():
        print(f'{key}:{value}')
```

```python
## Dictionary Comprehension

squares = {x:x**2 for x in range(5)}
print(squares)
```

```python
# Condition dictionary comprehension

evens = {x:x**2 for x in range(10) if x%2==0}
print(evens)
```

```python
## Practical Example

# Use a dictionary to count frequency of elements in list

numbers = [1,2,2,3,3,4,4,4,4]
frequency = {}

for number in numbers:
    if number in frequency:
        frequency[number] +=1
    else:
        frequency[number] =1
print(frequency)
```

```python
## Merge 2 dictionaries into one

dict1 = {'a':1,'b':2}
dict2 = {'b':3,'c':4}

merged_dict= {**dict1,**dict2}
print(merged_dict)
```
