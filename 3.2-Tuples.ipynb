Tuples are ordered collections of items that are immuatble. They are similar to lists, but their immutability makes them different.

```python
## Creating a tuples

empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst = list()
print(type(lst))
tp1 = tuple()
print(type(tp1))
```

```python
numbers = tuple([1,2,3,4,5,6])
numbers
```

```python
list((1,2,3,4,5,6))
```

```python
mixed_tuple = (1,'Hello World', 3.14, True)
print(mixed_tuple)
```

```python
## Accessing tuple Elemets
print(numbers)
print(numbers[0])
```

```python
## Tuple operations

concat_tuple = numbers + mixed_tuple
print(concat_tuple)
```

```python
## Immutable of Nature of Tuple

lst = [1,2,3,4,5]
print(lst)

lst[1]='Yuvraj'
print(lst)
```

```python
numbers[1] = 'Yuvraj'
```

```python
## Tuple methods

print(numbers.count(1))
print(numbers.index(3))
```

```python
## Packing and Unpacking Tuples

packed_tuple = 1,"Hello", 3.14
print(packed_tuple)
```

```python
a,b,c = packed_tuple
print(a)
print(b)
print(c)
```

```python
### Unpacking with *

numbers = (1,2,3,4,5,6)
first,*middle,last = numbers
print(first)
print(middle)
print(last)
```

```python
## Nested Tuples

lst = [[1,2,3,4],[6,7,8,9],[1,'Hello',3.14,'c']]
lst[0]
```

```python
lst[0][2]
```

```python
lst = [[1,2,3,4],[6,7,8,9],(1,'Hello',3.14,'c')]
lst[2][0:3]
```

```python
nested_tuple  = ((1,2,3),('a','b','c'),(True,False))

## access the elements inside tuple

print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## Iterating over nested tuples

for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()
```

Conclusion :

Tuples are versatile and useful in many real-world scenarios where an immutable and ordered collection of items is required. They are commonly used in data structures, function arguments amd return values, and as dictionary keys. Understanding how to leverage tuples effectively can improve the efficiency and readability of your Python code.
